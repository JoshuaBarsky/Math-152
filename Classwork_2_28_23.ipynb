```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
xlist = np.random.uniform(-10,10,1000)
```

```python
ylist = np.random.uniform(-10,10,1000)
```



---



alpha â‰¡ opacity/transparecy

c = color

s = size



---



```python
plt.figure(figsize=(6,6))
plt.scatter(xlist,ylist,alpha=0.5,c='GOLD',s=50)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.title('Go Slugs')
plt.show()
```

```python
clist = (xlist**2+ylist**2)
```



---


'clist' is the same as the list of squared radii (i.e. rlist**2).



---


```python
plt.figure(figsize=(6,6))
plt.scatter(xlist,ylist,alpha=0.5,c=clist,s=50)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.title('Go Slugs')
plt.show()
```
